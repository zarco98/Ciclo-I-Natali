```
adulto=input("¿Eres adulto? No:0, Si:1 ") 
hombre=input("¿Eres hombre? No:0, Si:1 ")
hijoUnico=input("¿Eres hijo único? No:0, Si:1 ")

if adulto=="1" and hombre=="1" and hijoUnico=="0":
  print("\nDebes prestar el servicio militar")
else:
  print("\nNo debes prestar el servicio militar")
```

```
#Casteando las variables

adulto=int(input("¿Eres adulto? No:0, Si:1 "))
hombre=int(input("¿Eres hombre? No:0, Si:1 "))
hijoUnico=int(input("¿Eres hijo único? No:0, Si:1 "))

if adulto==1 and hombre==1 and hijoUnico==0:
  print("\nDebes prestar el servicio militar")
else:
  print("\nNo debes prestar el servicio militar")
```

```
#Con booleanos (cambio de intérprete)

adulto=bool(int(input("¿Eres adulto? No:0, Si:1 ")))
hombre=bool(int(input("¿Eres hombre? No:0, Si:1 ")))
hijoUnico=bool(int(input("¿Eres hijo único? No:0, Si:1 ")))

servicioMilitar= (adulto==True) and (hombre==True) and (hijoUnico==False)
#Servicio militar sólo si adulto=True, hombre=True y hijoUnico=false
print("\nDebes prestar el servicio militar", servicioMilitar)
```

```
#Ejercicio 1

#Solicita la base del triángulo
base=float(input("\nPor favor ingresa la base del triángulo: "))
#Solicita la altura del triángulo
altura=float(input("\nPor favor ingresa la altura del triángulo: "))

#Cálculo del área de un triángulo: base*altura/2
area=(base*altura)/2
print("El área del triángulo es: ",area, "metros")
```

```
from numpy import pi
#Ejercicio 2

#Solicita el radio del círculo
radio=float(input("Por favor ingrese el radio del círculo: "))

#Cálculo de la circuferencia del círculo 2*pi*r
circunferencia=2*pi*radio
print("La circunferencia es: ",circunferencia)
```

```
#Ejercicio 3

import math

#Solicite en ángulo del triángulo
angulo=float(input("Por favor ingrese un ángulo: "))

#Cálculo del seno del ángulo
seno=math.sin(angulo)
print("El seno del ángulo es: ",seno)

#Cálculo del coseno del ángulo
coseno=math.cos(angulo)
print("El coseno del ángulo es: ",coseno)

#Cálculo de la tangente del ángulo
tangente=math.tan(angulo)
print("La tangente del ángulo es: ",tangente)
```
